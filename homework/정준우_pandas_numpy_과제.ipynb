#### birthDate 컬럼으로 age 컬럼 추가하기

##### 1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
import datetime as datetime

data_2['birthDate'] = pd.to_datetime(data_2['birthDate'])
data_2.head(30)
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
data_2['age'] = data_2['birthDate'].apply(lambda x: ( (pd.Timestamp.today() - x) / np.timedelta64(1, 'Y') ) )
data_2.age = data_2[data_2['age'].notnull()].age.astype('int')
data_2.head(30)
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
data_3 = data_2['age'].fillna(int(data_2.age.mean()))
data_2['age'] = data_3
data_2.tail()
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
data_2[['feet', 'inch']] = pd.DataFrame(data_2['height'].str.split('-', 2).tolist())
data_2.loc[:, 'heightCm'] = data_2.loc[:, 'feet': 'inch'].sum(axis = 1)
data_2 = data_2.drop('feet', axis=1)
data_2 = data_2.drop('inch', axis=1)
data_2.head(30)
```
